```python
# Download data
import os
import kagglehub
import pandas as pd

# Download latest version
path = kagglehub.dataset_download("govindaramsriram/sleep-time-prediction")

print("Path to dataset files:", path)

# Print file names and sizes in path using rich table
df = {"File": [], "Size(MB)": []}

for fn in os.listdir(path):
    if fn.endswith(".csv"):
        file_size = os.path.getsize(os.path.join(path, fn))
        df["File"].append(fn)
        df["Size(MB)"].append(f"{file_size / 1024 / 1024:.2f}")
df = pd.DataFrame(df)
display(df)
```

```python
# Load data
def load_data(path, fn):
    """
    Load data from path/fn
    """
    df = pd.read_csv(path + "/" + fn)
    print(f"Loaded {fn}")
    display(df.head())

    print("Metadata")
    display(df.info())
    
    print("Describe")
    display(df.describe())
    return df

df = load_data(path, "sleeptime_prediction_dataset.csv")
```
